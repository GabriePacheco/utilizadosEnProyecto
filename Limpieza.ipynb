```python
import pandas as pd 
import numpy as np 
import seaborn as sns 
import matplotlib.pyplot as plt 

import warnings
warnings.filterwarnings('ignore')
```

```python
targets = ["18+", "Adolescentes", "Amas", "Hombres18+", "Mujeres18+"]
years = [2023, 2022]
months = ["Enero", "Febrero", "Marzo", "Abril", "Mayo", "Junio", "Julio", "Agosto", "Septiembre", "Octubre", "Noviembre", "Diciembre"]

df_promocionales = pd.DataFrame()    

for year in years:
    for month in months:
        for target in targets: 
            
            file_name = f"Datos__{month}_{year}_{target}.xlsx"
            print(f"Cargando archivo: {file_name}")
            
            try: 
                data = pd.read_excel(file_name, sheet_name = "Sheet", header = 3 )
                data = data.iloc[0:-2]  
                data = data[['FECHA', 'VERSION', 'DURACION', 'TIPO', 'HORA',
                             'PROGRAMA', 'BREAK', 'TARIFA', 'CATEGORIA', 'ITEM',
                             'GENERO', 'SECCION/BLOQUE','CIUDAD', 'CAMPAÑA', 
                             'HOR_INI-HH:MM:SS','HOR_FIN-HH:MM:SS', 'Rat.Prom.Prog']]
                
                data.rename( columns = {'SECCION/BLOQUE': 'SECCION', 'HOR_INI-HH:MM:SS': "HORA_INICIO", 
                                        "HOR_FIN-HH:MM:SS": "HORA_FIN",'Rat.Prom.Prog': 'RATING' }, inplace = True)
                
                data["TARGET"] = target
                
                df_promocionales = pd.concat([df_promocionales, data ],  ignore_index = True )
                
                            
            except Exception as e: 
                print (f"ERROR EN EL ARCHIVO\n" )
                print(e)
                
                
            

```
