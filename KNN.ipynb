## CREACIÓ DE MODELOS 

### 1. OBTENCION DE LOS DATOS 

```python
# importa librerias necesarias
import pandas as pd 
import numpy as np 
```

```python
# cargar dataframe
df = pd.read_csv( 'promocionales.csv')
df.head(20)
```

```python
# limpiar y columnas 
df.drop(['Unnamed: 0', 'VERSION'], axis = 1, inplace = True )
```

```python
# renombrar columnas 
df.rename( columns =  { 'CAMPAÑA': 'CAMPANIA', 'Rat.Prom.Prog': 'RATING' }, inplace = True)
df.tail(19)
```

```python
df.describe()
```

### 2. SEPARAR EN TRAIN Y TEST

```python
# importar librerias necesarias 
from sklearn.model_selection import train_test_split
from sklearn.neighbors  import KNeighborsClassifier
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, confusion_matrix
```

```python
# separar datos en etiquetas y categorias 
X = df.drop("TIPO", axis = 1)
y = df['TIPO']
```

```python
#PREPOSESAR CARACTERISTICAS
encoders = {}
for column in X.select_dtypes(include = ['object'] ).columns:
    encoder = LabelEncoder()
    X[column] = encoder.fit_transform( X[column] )
    encoders[column] = encoder
    
```

```python
#separar en train y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.20 , random_state = 42 )
```

```python
X_test
```

### 3. ENTRENAR MODELO

```python
modelo = KNeighborsClassifier()
modelo.fit(X_train, y_train)
```

```python
y_pred = modelo.predict(X_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

matriz = confusion_matrix(y_test, y_pred)
class_labels = np.unique(np.concatenate((y_test, y_pred)))
plt.figure(figsize=(8, 6))

sns.heatmap(matriz, annot=True, fmt="d", cmap="Blues", xticklabels=class_labels, yticklabels=class_labels)
plt.xlabel("Predichos")
plt.ylabel("Prueba")
plt.title("Matriz de confucion KNeighborsClassifier")
plt.show()
```

```python
import pickle 

modelName  = "KNN_model.pkl"
encodersName = "KNN_encoder.pkl"

pickle.dump( modelo,  open(modelName, 'wb')) 
pickle.dump(encoders,  open(encodersName, 'wb')) 
```
