## CREACIÓ DE MODELOS 

### 1. OBTENCION DE LOS DATOS 

```python
# cargar dataframe
df = pd.read_csv( 'promocionales.csv')
df.head(3)
```

```python
# importa librerias necesarias
import pandas as pd 
import numpy as np 
```

```python
# limpiar y columnas 
df.drop(['Unnamed: 0', 'HORA_INICIO','HORA_FIN', 'HORA', 'FECHA', 'VERSION'  ], axis = 1, inplace = True )
```

```python
# renombrar columnas 
df.rename( columns =  {"SECCION/BLOQUE": 'SECCION', 'CAMPAÑA': 'CAMPANIA', 'Rat.Prom.Prog': 'RATING' }, inplace = True)
df.head(19)
```

### 2. SEPARAR EN TRAIN Y TEST

```python
# importar librerias necesarias 
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import r2_score, mean_absolute_error, mean_squared_error
```

```python
# separar datos en etiquetas y categorias 
X = df.drop("RATING", axis = 1)
y = df['RATING']
```

```python
#PREPOSESAR CARACTERISTICAS
encoders = {}
for column in X.select_dtypes(include = ['object'] ).columns:
    encoder = LabelEncoder()
    X[column] = encoder.fit_transform( X[column] )
    encoders[column] = encoder
    
```

```python
#separar en train y test
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size= 0.20 , random_state = 42 )
```

### 3. ENTRENAR MODELO

```python
modelo = DecisionTreeRegressor()
modelo.fit(X_train, y_train)
```

```python
y_pred = modelo.predict(X_test)
```

```python
r2_score(y_test, y_pred)
```

```python
mae = mean_absolute_error(y_test, y_pred)
mse = mean_squared_error(y_test, y_pred)
rmse = mean_squared_error(y_test, y_pred, squared=False)
r2 = r2_score(y_test, y_pred)

print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
print(f"R^2: {r2}")
```

```python
import pickle 

modelName  = "dtrm_model.pkl"
encodersName = "dtr_encoder.pkl"

pickle.dump( modelo,  open(modelName, 'wb')) 
pickle.dump(encoders,  open(encodersName, 'wb')) 
```
